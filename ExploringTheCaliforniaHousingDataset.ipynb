```python
fig, ax = plt.subplots(figsize = (12,8))  # 

plt.scatter(housing_data['total_rooms'],housing_data['median_house_value'])

plt.xlabel('Total rooms')
plt.ylabel('Median house values')
```

```python
fig, ax = plt.subplots(figsize = (12,8))  # 

plt.scatter(housing_data['housing_median_age'],housing_data['median_house_value'])

plt.xlabel('Median age')
plt.ylabel('Median house values')
```

```python
fig, ax = plt.subplots(figsize = (12,8))  # 

# to show in actual dollars
plt.scatter(housing_data['median_income'] * 10000,housing_data['median_house_value'])

plt.xlabel('Median income')
plt.ylabel('Median house values')
```

Clear upward flow -> clear positive relationship between median income and house prices

```python
# view correlation between columns
housing_data_corr = housing_data.corr()

housing_data_corr
```

If two variables move in the same direction together, they are positively correlated. Vice versa

```python
# Correlation are best visualized using heat map

fig, ax = plt.subplots(figsize=(12,10))

sns.heatmap(housing_data_corr, annot=True)
```

Colours indicate how positively/negatively correlated two features are.
Lighter colour indicate high positive correlation -> bright cream (1)
Correlation of -1 is almost black in colour
